#### 2.1 Phân tích RFM

RFM là một phương pháp được sử dụng để phân tích giá trị khách hàng, thường được sử dụng trong marketing cơ sở dữ liệu (kiểu như dựa vào dữ liệu về khách hàng để tiếp thị sản phẩm) và marketing trực tiếp và đã nhận được sự chú ý đặc biệt trong ngành bán lẻ và dịch vụ.

RFM định lượng giá trị của một khách hàng dựa trên 3 thông tin chính:

* Recency: Thời điểm gần nhất mà khách hàng đã mua hàng là bao lâu? Điều này cho biết mức độ hoạt động gần đây của khách hàng, trong đó chỉ số lớn hơn thể hiện xu hướng rời bỏ cao. Đây là một cảnh báo cho doanh nghiệp để điều chỉnh sản phẩm hoặc chính sách để đáp ứng nhu cầu của khách hàng hoặc cải thiện chất lượng dịch vụ.
* Frequency: Tần suất mua hàng của khách hàng. Nếu khách hàng mua hàng thường xuyên, doanh số bán của công ty cũng cao và giá trị của họ tăng lên. Tuy nhiên, chỉ dựa trên tần suất mua hàng không đánh giá đầy đủ ảnh hưởng lên doanh thu, vì giá trị đơn hàng cũng là yếu tố quan trọng cho biết khách hàng tiềm năng là như thế nào.
* Monetary: Tổng số tiền khách hàng đã chi tiêu. Đây là yếu tố quan trọng nhất ảnh hưởng đến doanh số bán. Cụ thể, doanh nghiệp quan tâm đến việc khách hàng đã chi bao nhiêu tiền để mua sản phẩm của họ. Monetary ảnh hưởng trực tiếp đến doanh thu và giữa đó còn được tác động gián tiếp thông qua Recency và Frequency.

Sau khi thu thập thông tin từ ba yếu tố trên, người phân tích có thể sử dụng thuật toán K-Mean clustering, một thuật toán học không giám sát để nhóm các khách hàng có cùng mức độ quan trọng thành các nhóm khác nhau.

Phân khúc khách hàng dựa trên RFM có thể giúp đánh giá hành trình khách hàng (customer journey) dựa trên nhiều yếu tố, từ nhiều góc. Từ đó, phân tích RFM giúp bạn xây dựng các thông điệp, chiến dịch và chương trình ưu đãi khác nhau cho từng nhóm khách hàng, dựa trên hành vi tiêu dùng và mối quan hệ của họ với công ty. Mỗi chỉ số RFM này đã được chứng minh là có hiệu quả trong việc dự đoán hành vi của khách hàng trong tương lai và tăng doanh thu. Những khách hàng đã mua hàng trong quá khứ gần đây có nhiều khả năng sẽ làm như vậy trong tương lai gần.

Dựa vào đây, các hoạt động marketing có thể tăng tỷ lệ phản hồi, tăng khả năng giữ chân khách hàng (retention rate), sự hài lòng của khách hàng và giá trị vòng đời của khách hàng (CLTV), phân khúc khách hàng cho giới thiệu sản phẩm mới, tăng sự trung thành và tăng tương tác với khách hàng, tối ưu chi phí marketing và cải thiện ROI.

### 2.2 Thuật toán phân cụm K-means (K-means Clustering
K-means clustering là một trong những thuật toán học máy không giám sát (unsupervised learning) đơn giản và phổ biến.

Thông thường, các thuật toán không giám sát đưa ra các kết luận từ các tập dữ liệu chỉ sử dụng các vectơ đầu vào mà không gắn nhãn dữ liệu hoặc giả định trước kết quả.

Mục tiêu của K-means rất đơn giản: nhóm các điểm dữ liệu tương tự lại với nhau và khám phá các cấu trúc (pattern) cơ bản. Để đạt được mục tiêu này, K-means tìm kiếm một số lượng cố định (k) các cụm trong một tập dữ liệu.
Cụm là một tập hợp các điểm dữ liệu được tập hợp lại với nhau vì có những điểm tương đồng nhất định.

Bạn sẽ xác định số mục tiêu k, là số lượng điểm trung tâm (centroid) bạn cần trong tập dữ liệu. Centroid trong toán học là trọng tâm của một hình, ví dụ trọng tâm của tam giác. Trong thuật toán K-means, Centroid là vị trí thực hoặc ảo đại diện cho trung tâm của cụm.
Mọi điểm dữ liệu được phân bổ cho từng cụm sao cho tổng bình phương khoảng cách từ các điểm dữ liệu trong cụm tới các điểm trung tâm Centroid là nhỏ nhất.
 ‘Means’ trong K-means đề cập đến điểm trung bình của dữ liệu; hay nói cách khác là tìm ra điểm trung tâm.

Nói cách khác, thuật toán K-means xác định k số Centroid, và sau đó phân bổ mọi điểm dữ liệu vào các cụm gần nhất, mỗi cụm tương ứng với 1 điểm trung tâm Centroid.

```python
#Sử dụng hàm để check dtype, số lượng giá trị unique, null và duplicated của từng cột trong tập dữ liệu
def check(df):
    list = []
    for col in df.columns:
        Dtype = df[col].dtype
        Unique_value = df[col].nunique()
        Null = df[col].isnull().sum()
        Duplicated = df[col].duplicated().sum()
        list.append([col,Dtype,Unique_value,Null,Duplicated])
        bang_check = pd.DataFrame(list, columns =["Column Name","Dtype","Unique","Null","Duplicated"])
    return bang_check

check(df)
```

### 6. Xử lý Outlier

```python
# Tạo hàm để xử lý outlier
def handle_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].apply(lambda x: upper_bound if x > upper_bound else lower_bound if x < lower_bound else x)
    return data
#Xử lý oulier cột CustAccountBalance
handle_outliers_iqr(clean_data,"CustAccountBalance")
#Xử lý outlier cột TransactionAmount (INR)
handle_outliers_iqr(clean_data,"TransactionAmount (INR)")
```

### 2. Xây dựng model

#### Elbow-Curve

Phương pháp Elbow là một cách giúp ta lựa chọn được số lượng các cụm phù hợp dựa vào đồ thị trực quan hoá bằng cách nhìn vào sự suy giảm của hàm biến dạng và lựa chọn ra điểm khuỷ tay (Elbow Point). Để tìm hiểu phương pháp Elbow, bên dưới chúng ta cùng thử nghiệm vẽ biểu đồ hàm biến dạng bằng cách điều chỉnh số lượng cụm của thuật toán K-Means.

```python
# Elbow-curve
kmeans_set = {"init":"random", "max_iter":300, "random_state":42}
ssd = []
range_n_clusters = range(2,12)
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters,**kmeans_set)
    kmeans.fit(data_scaled)

    ssd.append(kmeans.inertia_)

# Vẽ biểu đồ
ig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(y=ssd, x=range_n_clusters, marker='o',ax=ax )
plt.title('The Elbow Method')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show();
```

#### Sihoutte Score

Silhouette Score là một công cụ để đánh giá tính phù hợp của kết quả phân cụm bằng cách cung cấp thước đo định lượng về mức độ phân biệt và xác định rõ ràng của các cụm. Silhouette Score định lượng mức độ phù hợp của một điểm dữ liệu với cụm được chỉ định và mức độ khác biệt của nó với các cụm khác. Nó đo lường sự gắn kết và phân tách các điểm dữ liệu trong các cụm và giúp xác định xem các cụm có được phân tách tốt và đồng nhất bên trong hay không. Ở đây, chúng em sẽ tính Sihoutte Score cho số lượng cụm k từ 2 đến 12 và biểu diễn trên biểu đồ đường để đánh giá.

```python
# Sihoutte Score
silhouette_scores=[]
k_range= range(2,12)

for i in k_range:
    kmeans = KMeans(n_clusters= i,**kmeans_set)
    kmeans.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Biểu diễn Sihoutte Score
fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(x=k_range, y=silhouette_scores, ax=ax )
plt.title('The silhouette Scores Method for Kmeans Algorithem')
plt.xlabel('K')
plt.ylabel('Silhouette scores')
plt.show()
```

#### Dendogram (Hierachical Clustering)

Phương pháp phân cụm phân cấp (Hierachical Clustering) không yêu cầu khai báo trước số lượng cụm, được biểu diễn qua đồ thị Dendogram (Cây quyết định).

```python
# Dendogram
dendrogram = sch.dendrogram(sch.linkage(data_scaled, method='ward'))

# Display the dendrogram
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

#### K-means Clustering

Từ việc sử dụng 2 phương pháp là Sihoutte Score và Elbow Curve và phân cụm phân cấp qua đồ thị Dendogram, chúng em tìm ra số lượng cụm tối ưu nhất là k=4 và quyết định phân chia khách hàng thành 4 cụm để tiếp tục với thuật toán phân cụm K-means K-means (K-means Clustering)
